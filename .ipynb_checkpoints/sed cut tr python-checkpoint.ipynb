### Ranking Vertical Delfos

**uso de**: grep cut sort uniq tr sed

```python
!cat '/home/marco/Descargas/Analytics Vista de Usuarios Vuelos por agencia sin DELFOS 20210125-20210129.csv' | sed -n '8,3869p' | tr '/|=|?' ',' | grep delfos | cut -d, -f4 | sort | uniq -c | sort -r
```

### Ranking Vertical Agencias


```python
!cat '/home/marco/Descargas/Analytics Vista de Usuarios Vuelos por agencia sin DELFOS 20210125-20210129.csv' | sed -n '8,3869p' | tr '/|=|?' ',' | grep -v delfos | cut -d, -f4 | sort | uniq -c | sort -r
```

Del archivo 14dec2020.csv extraemos con **cut** las columnas de la 3 a la 12

```python
!cat '/home/marco/Descargas/Analytics Vista de Usuarios Vuelos por agencia sin DELFOS 20210125-20210129.csv' | sed -n '8,3869p' | tr '/|=|?' ',' > resultados.csv
```

```python
!cat resultados.csv | cut -d',' -f3-12 | wc -l
```

Del archivo resultante, convertimos todos ciertos caracteres a ',' para separar bien los campos

```python
!cat resultados.csv | tr '/|=|?' ',' | cut -d ',' -f3-10 > resultados_filtrados.csv
```

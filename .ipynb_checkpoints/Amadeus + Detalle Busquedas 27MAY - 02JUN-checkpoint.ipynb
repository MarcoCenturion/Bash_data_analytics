## Comandos Master Pricer

|Busqueda|Comando|
|--------|-------|
|SIMPLE: 2 pasajeros COR MIA COR sin filtros|FXD2COR/D10AUGMIA/D20AUGCOR|
|Idem anterior + 2 APTS origen (pasajero de LA PAMPA)|FXD2COR,BUE,MDZ/D10AUGMIA/D20AUGCOR,MDZ,BUE|
|idem + 2 apts destino (Vacacional que busca CUN, pero que podemos ganar la venta por ofrecerle un destino de playa similar)|FXD2COR/D10AUGCUN,PUJ,SDQ/D20AUGCOR|
|idem `excluyendo` AR|FXD2COR/D10AUGCUN,PUJ,SDQ/D20AUGCOR/`A-AR`|
|idem `forzando que solo traiga vuelos de AR` (hasta 3 cías)|FXD2COR/D10AUGMIA,FLL,TPA/D20AUGCOR//`AAR`|
|La ruta mas barata a estos 3 destinos, con AR, CM o AV|FXD2COR/D10AUGMIA,FLL,TPA/D20AUGCOR//`AAR,CM,AV`|
|La mejor tarifa a cualquiera de estos 3 destinos, NAT BPS ó SSA incluyendo netas, publicas, nego G3, nego JJ y nego AD|FXD2COR/D10AUGBPS,SA,NAT/D20AUGCOR/R,UP,U`*ARBR00,U*AZUL,U*OPERINT5`|
|La mas barata a PUJ SDQ o HAV publicas, netas y nego CM|FXD2COR/D10AUGPUJ,SDQ,HAV/D20AUGCOR/`RPFA,U*ARVPU,UP`|
|Radio de 120 KM alrededor de un TRN|FXD2COR/D10AUGTRN/`DI-120K`/D20AUGCOR|

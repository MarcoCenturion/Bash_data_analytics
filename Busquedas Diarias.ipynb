## Armador de la imagen amigable al usuario

```python
aptorigen = input("Ingresar el aeropuerto de Origen: ")

# DEFINIR LA VARIABLE A GRAFICAR
# vuelos.loc[vuelos['Tramo'].str.startswith('/COR')][['Agencia','Tramo','Fecha']]
# vuelos.head(20)[['Agencia','Tramo','Ida','Regreso','Número de visitas a páginas']]
# vuelos.loc[vuelos['Tramo'].str.startswith('/JUJ')][['Agencia','Tramo','Fecha']]
# clientes.loc[clientes['Cliente'].str.contains('MATIEN'),['Sucursal','Cliente','Sector','may-18','jun-18','jul-18','ago-18','sep-18','Suma']] 

VARIABLE=vuelos.loc[vuelos['Tramo'].str.startswith('/'+aptorigen)][['Agencia','Tramo','Fecha']]

def render_mpl_table(data, col_width=45.0, row_height=0.500, font_size=9
                     ,
                     header_color='#F78014', row_colors=['#f9d004', 'w'], edge_color='w',
                     bbox=[0, 0, 1, 1], header_columns=0,
                     ax=None, **kwargs):
    if ax is None:
        size = (np.array(data.shape[::-1]) + np.array([0, 1])) * np.array([col_width, row_height])
        fig, ax = plt.subplots(figsize=size)
        ax.axis('off')

    mpl_table = ax.table(cellText=data.values, bbox=bbox, colLabels=data.columns, **kwargs)

    mpl_table.auto_set_font_size(False)
    mpl_table.set_fontsize(font_size)

    for k, cell in  six.iteritems(mpl_table._cells):
        cell.set_edgecolor(edge_color)
        if k[0] == 0 or k[1] < header_columns:
            cell.set_text_props(weight='bold', color='w')
            cell.set_facecolor(header_color)
        else:
            cell.set_facecolor(row_colors[k[0]%len(row_colors) ])
    return ax

render_mpl_table(VARIABLE, header_columns=0, col_width=2.0)
```

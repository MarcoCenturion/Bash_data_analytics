```python
# stock
stock_ars = 5123000
stock_usd = 3222
stock_eur = 2100
stock_brl = 11300
stock_che = 2220000

# conversion
compra_usd = 298
venta_usd = 295
compra_eur = 250
venta_eur = 240 
compra_brl = 34
venta_brl = 32
compra_ch = 1.43
```

```python
balance = [stock_ars,stock_usd*venta_usd,stock_eur*venta_eur,stock_brl*venta_brl,stock_che]
```

```python
labels = 'Pesos','Dolares','Euros','Reales','Cheques'
fig1, ax1 = plt.subplots(figsize=(10,10))
ax1.pie(balance, labels=labels, autopct='%1.3f%%', startangle=90)

ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,4))

ax.bar(labels, balance, label=labels)

ax.set_ylabel('Millones de ARS', fontsize = 22)
ax.set_title('Activos en stock en tiempo real', fontsize = 20)
ax.legend(bbox_to_anchor=(1, 0.6), fontsize=16) #ubicacin leyenda
ax.legend(title='Monedas')

plt.show()
```

```python
ax.set_title('titulo', fontsize=25)  # Titulo de la grafica
ax.set_ylabel('titulo', fontsize=20) # Titulo eje y
ax.set_xlabel('titulo', fontsize=20) # Titulo eje x

```
